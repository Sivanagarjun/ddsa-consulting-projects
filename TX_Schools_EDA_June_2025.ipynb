```python
# Code snippet 1
# Setting the dataset URL
url = 'https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/2023%20Texas%20High%20Schools%20Scores.csv'

# Display the url
url
```

```python
# Code snippet 2
# Importing dataset using pandas
import pandas as pd

# Importing the dataset and assigning it to a variable
data = pd.read_csv(url)

# Display the data
data
```

```python
# Code snippet 3
# Display the size of the dataset
data.shape
```

```python
# Code snippet 4
# Display the column names of the dataset
data.columns
```

```python
# Code snippet 5
# Display the statistical measures of the dataset
data.describe()
```

```python
# Code snippet 9
# Ignoring/dropping 'Home Prices Ranking', 'Community Health Ranking' variables because they are misisng more than 20% data
# Quantitative data preparation
data80 = data.drop(['Home Prices Ranking', 'Community Health Ranking'], axis=1)

# Display the data80 columns
data80.columns
```

```python
# Code snippet 10
# Displaying the size of data80
data80.shape
```

```python
# Code snippet 11
# Dropping 'School', 'School URL',  'District URL','City URL',   'Phone' as noise variables
# Qualitative data preparation
predictors = data80.drop(['School', 'School URL',  'District URL','City URL', 'Phone'], axis=1)

# Displaying the columns of predictors
predictors.columns
```

```python
# Code snippet 12
# Setting Zip as categorical
predictors['Zip'] = predictors['Zip'].astype('category')
```

```python
# Code snippet 13
# Installing ydata-profiling package for detailed EDA/data profiling
!pip install ydata-profiling
```

```python
# Code snippet 14
# Importing ProfileReport from ydata-profiling package
from ydata_profiling import ProfileReport
```

```python
# Code snippet 15
# Generating detailed Profile report using ProfileReport from ydata-profiling package
profile = ProfileReport(predictors)

# Converting profile report to an interactive html report
profile.to_file("TX Schools 2025 Profile Report.html")
```
