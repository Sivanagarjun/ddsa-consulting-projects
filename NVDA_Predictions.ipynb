```python
# cs 2
# Importing Prophet and yahoo finance packages
from prophet import Prophet
import yfinance as yf
```

```python
# Cs 3
# Set start dates and end dates
start_date = '2015-01-01'
end_date = '2025-07-19'

ticker = 'NVDA'
```

```python
# cs 4
# Importing the stock data
data = yf.download(ticker, start_date, end_date)

data
```

```python
# cs 5
# Resetting the index
data.reset_index(inplace=True)

data
```

```python
# Cs 6
# Selecting date and adjusted close
data = data[['Date', 'Close']]

data
```

```python
# Cs 7
# Renaming date as ds and adj close as y
data.columns = ['ds', 'y']

data
```

```python
# Code snippet 8
# Initialize Prophet
model = Prophet()

# Code snippet
# Training the model with the date and y values
model.fit(data)

# Code snippet
# Setting a future dataframe
future = model.make_future_dataframe(periods=30)

# Code snippet
# Predicting the incidents for future period
forecast = model.predict(future)

# Code snippet
# Creating a subset with predictions, lower and upper values
forecast_subset = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
# Code snippet 8.1
# Creating a final dataset with actuals, predictions, lower and upper values
import pandas as pd
# Ensure 'ds' columns in both dataframes are datetime type before merging
data['ds'] = pd.to_datetime(data['ds'])
forecast_subset['ds'] = pd.to_datetime(forecast_subset['ds'])

# Merge on 'ds' column
forecast_subset_merged = pd.merge(data, forecast_subset[['ds', 'yhat', 'yhat_lower', 'yhat_upper']], on='ds', how='outer')
forecast_subset_merged
```

```python
# Code snippet 9
# Exporting the predictions
forecast_subset_merged.to_csv('NVDA Stock Price predictions.csv')
```
