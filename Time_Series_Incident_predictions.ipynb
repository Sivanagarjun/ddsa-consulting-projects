```python
# Code snippet 1
# Importing the time series data
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/Statefarm%20Incidents.csv')

data
```

```python
# Code snippet 2
# Importing Prophet package for time series regression ML modeling
from prophet import Prophet
```

```python
# Code snippet 3
# Renaming Date as ds and Incidents as y
data.columns = ['ds', 'y']

data
```

```python
# Code snippet 4
# Initialize Prophet
model = Prophet()
```

```python
# Code snippet 5
# Training the model with the date and y values
model.fit(data)
```

```python
# Code snippet 6
# Setting a future dataframe
future = model.make_future_dataframe(periods=30)
```

```python
# Code snippet 7
# Predicting the incidents for future period
forecast = model.predict(future)
```

```python
# Code snippet 8
# Displaying the predictions
forecast
```

```python
# Code snippet 9
# Creating a subset with predictions, lower and upper values
forecast_subset = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]


forecast_subset
```

```python
# Code snippet 9.1
# Creating a final dataset with actuals, predictions, lower and upper values

# Ensure 'ds' columns in both dataframes are datetime type before merging
data['ds'] = pd.to_datetime(data['ds'])
forecast_subset['ds'] = pd.to_datetime(forecast_subset['ds'])

# Merge on 'ds' column
forecast_subset_merged = pd.merge(data, forecast_subset[['ds', 'yhat', 'yhat_lower', 'yhat_upper']], on='ds', how='outer')
forecast_subset_merged
```

```python
# Code snippet 10
# Exporting the predictions
forecast_subset_merged.to_csv('Statefarm incidents forecast subset.csv')
```
