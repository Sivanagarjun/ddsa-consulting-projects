```python
# Code snippet 2
# Importing the dataset
import pandas as pd
url = 'https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/2023%20Texas%20High%20Schools%20Scores.csv'

data = pd.read_csv(url)

data
```

```python
# Code snippet 2
# Displaying data columns
data.columns
```

```python
# Code snippet 3
# Setting the target variable
y = 'Standard Score'
```

```python
# Code snippet 4
# Setting variables to ignore(noise/redundant/feeder variables and ones that are missing more than 20% data)
ignored_cols = ['School', 'School URL', 'District URL',  'City URL',
       'Phone', 'Home Prices Ranking', 'Community Health Ranking']
```

```python
# Code snippet 5
# Setting the numerical columns
num_cols = ['Number Students', 'Number Full-time Teachers',
       'Student/Teacher Ratio', 'Percent Free/Disc Lunch',
       'Percent Homes Rented', 'Percent of Unemployment',
       'Percent Parents with Masters and above education',
       'Percent parents with less than High School education',
       'Percent Home Incomes with less than Median income']
```

```python
# Code snippet 6
# Setting the categorical columns
cat_cols = ['District', 'City', 'Zip', 'County', 'Is Title I', 'Is Charter', 'Is Magnet',
       'Is Virtual']
```

```python
# Code snippet 7
# Importing Regression components from pycaret
from pycaret.regression import *
```

```python
# Code snippet 8
# Setting /configuring pycaret ML modeling
regression_setup = setup(data,
                         target = y,
                         ignore_features = ignored_cols,
                         numeric_features = num_cols,
                         categorical_features = cat_cols)
```

```python
# Code snippet 9
# Invoking ML algorithms
compare_models()
```

```python
# Code snippet 10
# Build the ML model using the best algo
ml_model = create_model('huber')
```

```python
# Code snippet 11
# Predicting standard scores using the best model
score_predictions = predict_model(ml_model, data)

# Exporting predictions to a csv file
score_predictions.to_csv('TX Standard Scores Predictions.csv')
```
