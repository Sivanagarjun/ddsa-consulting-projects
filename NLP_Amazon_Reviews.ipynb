```python
# Code snippet 1
# Importing the data(unstructued text data)
url = 'https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/Amazon%20reviews.txt'

# importing requests to import unstructured dataset
import requests

# Importing the reviews from the URL
reviews = requests.get(url)

# Extracting text from the reviews
reviews_text = reviews.text

# Displaying the reviews
reviews_text
```

```python
# Code snippet 2
# Importing Sentiment analysis packages
import nltk  # Natural Language Processing Tool Kit

# Importing sentiment anlaysis component
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Importing latest dictionary
nltk.download('vader_lexicon')
```

```python
# Code snippet 3
# Initialize/start the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Code snippet 4
# Analyzing the sentiment scores
sentiment_scores = sia.polarity_scores(reviews_text)

# Display the sentiment scores
sentiment_scores
```

```python
# Code snippet 4.1. - Unit testing
# Analyzing the sentiment scores

sample_text = "Product is ok but expensive :( "
sentiment_scores1 = sia.polarity_scores(sample_text)

# Display the sentiment scores
sentiment_scores1
```
