```python
# Code snippet 1
# Importing the dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/Zillow%20Tarrant%20County%20Raw%20Data.csv')

# Display the data
data
```

```python
# Code snippet 2
# Importing clustering ML components
from pycaret.clustering import *
```

```python
# Code snippet 3
# Setting up pycaret configuration
cluster_model = setup(data)
```

```python
# Code snippet 4
# Create a clustering model using k-means ML algo
kmeans_model= create_model('kmeans')
```

```python
# Code snippet 5
# Assigning the clusters to the datapoints
kmeans_clusters = assign_model(kmeans_model)

# Display the clusters
kmeans_clusters
```

```python
# Code snippet #6
# Display the cluster sizes
kmeans_clusters['Cluster'].value_counts()
```
