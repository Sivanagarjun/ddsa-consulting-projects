```python
# Code snippet 5
# Setting the dataset URL
url = 'https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/2023%20Texas%20High%20Schools%20Scores.csv'

# Importing dataset using pandas
import pandas as pd

# Importing the dataset and assigning it to a variable
data = pd.read_csv(url)

# Dropping not needed variables (after data preparation)
predictors = data.drop(['School', 'School URL',  'District URL','City URL', 'Phone', 'Home Prices Ranking', 'Community Health Ranking'], axis=1)

# Displaying the columns of predictors
predictors.columns
```

```python
# Code snippet 6
# Generating detailed EDA report using sweetviz package
report = sv.analyze(predictors)

# Converting report into an interactive HTML report
report.show_html("TX High Schools - Sweetviz EDA report.html")
```
