```python
# Code snippet 2 - importing the dataset
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/Trivalley%20Customer%20Spend%20and%20Buyer%20Flag%20Data.csv')
df.shape
```

```python
# Code snippet 3 - displaying the columns
df.columns
```

```python
# Code snippet 4 - creating predictors variable
pred_columns = ['AveMonthSpend', 'BikeBuyer',
       'City', 'StateProvinceName', 'CountryRegionName', 'PostalCode',
       'Education', 'Occupation', 'Gender',
       'MaritalStatus', 'HomeOwnerFlag', 'NumberCarsOwned',
       'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome']
# Create a predictors dataset
df1 = df[pred_columns]

# Displaying the shape of the dataframe
df1.shape
```

```python
# Code snippet 5 - using sketch to display a bar chart of Gender
df1.sketch.howto('create a barchart of gender')
```

```python
# Code snippet 5.1 - to display a bar chart of Gender

# import matplotlib library
import matplotlib.pyplot as plt

# create a barchart of gender
df1['Gender'].value_counts().plot(kind='bar')

# set title and labels
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

# show plot
plt.show()
```

```python
# Code snippet 6.1 - To find the missing values in the dataset

for col in df1.columns:
    missing_values_percentage = (df1[col].isnull().sum()/len(df1))*100
    print('Feature: {}, Missing Values Percentage: {:.2f}%'.format(col, missing_values_percentage))
```

```python
# Code snippet 12 - Identify numerical features with less than 20% missing data and replace the missing data with mean
df1.sketch.howto('Identify numerical features with less than 20% missing data and replace the missing data with mean')
```

```python
# Code snippet 12.1 -  To Identify numerical features with less than 20% missing data and replace the missing data with mean

# Get list of numerical columns
numerical_columns = df1.select_dtypes(include=['int64', 'float64']).columns

# Iterate through numerical columns and identify those with less than 20% missing data
for col in numerical_columns:
    # Calculate percentage of missing data
    missing_data_percentage = df1[col].isnull().sum() / df1.shape[0] * 100

    # If missing data is less than 20%, replace with mean
    if missing_data_percentage < 20:
        df1[col] = df1[col].fillna(df1[col].mean())
```
