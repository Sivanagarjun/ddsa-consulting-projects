```python
# Code snippet 1
# importing pandas and other relevant packages
import pandas as pd
import numpy as np

# Importing SQLlite 3 package (to run SQL queries)
import sqlite3
```

```python
# Code snippet 2
# Importing the dataset
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/refs/heads/main/Tarrant%20County%20Income%20Dataset%20SQL%20CSV.csv')

data
```

```python
# Code snippet 3
# Creating a sample (dummy) database as a connection
conn = sqlite3.connect('sample.db')
```

```python
# Code snippet 4
# Storing the pandas dataframe db as a table in the sample database
data.to_sql('df_sql', conn)
```

```python
# Code snippet 5
# Selecting all the data from the table
sql_query = "SELECT * FROM df_sql"
df1 = pd.read_sql(sql_query, conn)
df1
```

```python
# Code snippet 6
# Selecting all the data from the table where age > 50
sql_query = "SELECT * FROM df_sql WHERE age > 50"
df1 = pd.read_sql(sql_query, conn)
df1
```

```python
# Code snippet 7
# Selecting all the data from the table where age <= 50
sql_query = "SELECT * FROM df_sql WHERE age <= 50"
df1 = pd.read_sql(sql_query, conn)
df1
```

```python
# Code snippet 8
# Selecting all the data from the table where age <= 30 and income > 50k
sql_query = ''' SELECT * FROM df_sql WHERE age <= 30 and income = ">50k" '''
df1 = pd.read_sql(sql_query, conn)
df1
```

```python
# Code snippet 9
# Selecting all the distinct marital statuses
sql_query = ''' SELECT distinct(Marital_Status) FROM df_sql  '''
df1 = pd.read_sql(sql_query, conn)
df1
```

```python
# Code snippet 10
# Selecting the marital status and the respective counts, order by count descending
sql_query = ''' SELECT Marital_Status, count(*) FROM df_sql group by Marital_Status order by count(*) desc'''
df1 = pd.read_sql(sql_query, conn)
df1
```

```python
# Code snippet 11
# Selecting the education and the respective counts, order by count descending
sql_query = ''' SELECT education, count(*) FROM df_sql group by education order by count(*) desc'''
df1 = pd.read_sql(sql_query, conn)
df1
```

```python
# Code snippet 12
# Selecting the education and the respective counts (with atleast 1000 records), order by count descending
sql_query = ''' SELECT education, count(*) FROM df_sql group by education having count(*) > 1000 order by count(*) desc'''
df1 = pd.read_sql(sql_query, conn)
df1
```

```python
# Code snippet 13
# Selecting the Occupation and the respective counts, order by count descending
sql_query = ''' SELECT occupation, count(*) FROM df_sql group by occupation order by count(*) desc'''
df1 = pd.read_sql(sql_query, conn)
df1
```
